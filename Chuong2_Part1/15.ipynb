```python
import math

m = float(input("Nhập khối lượng nước (gram): "))
dT = float(input("Nhập độ biến thiên nhiệt độ ΔT (°C): "))

C = 4.186        
JOULE_TO_KWH = 2.777e-7
COST_PER_KWH_CENTS = 8.9  

Q_joules = m * C * dT

Q_kwh = Q_joules * JOULE_TO_KWH
cost_cents = Q_kwh * COST_PER_KWH_CENTS

print("Năng lượng cần (J):", Q_joules)
print("Năng lượng tương đương (kWh):", Q_kwh)
print("Chi phí (cent):", cost_cents)
```
