```python
from math import *
def NT (x):
    if x  <= 1:
        return False
    for i in range (2, int(sqrt(x)) +1):
        if x% i ==0:
            return False
    return True
num = int(input("Nhap mot so can kiem tra:"))
if NT(num):
    print (num," La so nguyen to:")
else:
    print (num," Khong phai la so nguyen to:")
    
```
