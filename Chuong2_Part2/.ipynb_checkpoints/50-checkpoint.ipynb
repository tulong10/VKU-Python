```python
# Bai 53
def find_all_subsets_manual(nums):
  subsets = [[]]
  for num in nums:
    new_subsets = [current_subset + [num] for current_subset in subsets]
    subsets.extend(new_subsets)
  return subsets
my_list = [1, 2, 3]
subsets_manual = find_all_subsets_manual(my_list)
print(f"\nCác danh sách con của {my_list} là:")
print(subsets_manual)
```

```python
# Bai 54
def format_list_with_and(items):
  item_count = len(items)

  if item_count == 0:
    return ""
  if item_count == 1:
    return items[0]

  if item_count == 2:
    return f"{items[0]} and {items[1]}"
  
  # Nối tất cả các phần tử trừ phần tử cuối bằng dấu phẩy
  all_but_last = ', '.join(items[:-1])
  last_item = items[-1]
  
  return f"{all_but_last} and {last_item}"

word_list = []

print("Nhập các từ, mỗi từ trên một dòng.")

while True:
  word = input()
  if word == "": 
    break       
  word_list.append(word)
formatted_result = format_list_with_and(word_list)

print("\nChuỗi đã được định dạng là:")
print(formatted_result)
```

```python
# Bai 55
import zipfile
import os

def nen_tep(danh_sach_tep, ten_tep_nen):
    print(f"Bắt đầu nén vào '{ten_tep_nen}'...")
    # Mở tệp ZIP ở chế độ 'w' (write - ghi)
    with zipfile.ZipFile(ten_tep_nen, 'w', zipfile.ZIP_DEFLATED) as zipf:
        for tep in danh_sach_tep:
            # Kiểm tra xem tệp có tồn tại không trước khi thêm
            if os.path.exists(tep):
                # Ghi tệp vào kho lưu trữ ZIP
                # arcname=os.path.basename(tep) giúp loại bỏ cấu trúc thư mục
                zipf.write(tep, arcname=os.path.basename(tep))
                print(f"  -> Đã thêm: {tep}")
            else:
                print(f"  -> Cảnh báo: Tệp '{tep}' không tồn tại, đã bỏ qua.")
    print(f" Nén thành công tệp '{ten_tep_nen}'")

def giai_nen_tep(ten_tep_nen, thu_muc_giai_nen):
 
    # Kiểm tra xem tệp ZIP có tồn tại không
    if not os.path.exists(ten_tep_nen):
        print(f"Tệp nén '{ten_tep_nen}' không tồn tại.")
        return

    print(f"Bắt đầu giải nén '{ten_tep_nen}' vào '{thu_muc_giai_nen}'...")
    os.makedirs(thu_muc_giai_nen, exist_ok=True)
    
    with zipfile.ZipFile(ten_tep_nen, 'r') as zipf:
        # Trích xuất tất cả nội dung
        zipf.extractall(thu_muc_giai_nen)
    
    print(f"Giải nén thành công vào thư mục '{thu_muc_giai_nen}'")

# --- Chương trình chính có menu lựa chọn ---
if __name__ == "__main__":
    while True:
        print("\n--- CHƯƠNG TRÌNH NÉN & GIẢI NÉN TỆP ---")
        print("1. Nén các tệp")
        print("2. Giải nén một tệp")
        print("3. Thoát")
        
        lua_chon = input("Nhập lựa chọn của bạn (1, 2 hoặc 3): ")

        if lua_chon == '1':
            tep_dau_ra = input("Nhập tên tệp nén đầu ra (ví dụ: archive.zip): ")
            danh_sach_can_nen = []
            print("Nhập tên các tệp cần nén (nhấn Enter ở dòng trống để kết thúc):")
            while True:
                tep = input()
                if tep == "":
                    break
                danh_sach_can_nen.append(tep)
            
            if danh_sach_can_nen:
                nen_tep(danh_sach_can_nen, tep_dau_ra)
            else:
                print("Không có tệp nào được chọn để nén.")

        elif lua_chon == '2':
            tep_can_giai_nen = input("Nhập tên tệp cần giải nén (ví dụ: archive.zip): ")
            thu_muc_dau_ra = input("Nhập tên thư mục để giải nén vào (ví dụ: extracted_files): ")
            giai_nen_tep(tep_can_giai_nen, thu_muc_dau_ra)

        elif lua_chon == '3':
            print("Thoát")
            break
        else:
            print("Lựa chọn không hợp lệ, vui lòng thử lại.")
```
