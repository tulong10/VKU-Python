```python
def giai_ma_caesar(tin_nhan):
    ket_qua = ""
    do_dich_chuyen = 3
    for ky_tu in tin_nhan:
        if ky_tu.isupper():  
            # Chuyển ký tự thành mã ASCII, dịch chuyển và quay vòng trong bảng chữ cái
            ma_ky_tu = ord(ky_tu)
            ma_ky_tu_moi = (ma_ky_tu - ord('A') - do_dich_chuyen) % 26 + ord('A')
            ket_qua += chr(ma_ky_tu_moi)
        elif ky_tu.islower():  
            ma_ky_tu = ord(ky_tu)
            ma_ky_tu_moi = (ma_ky_tu - ord('a') - do_dich_chuyen) % 26 + ord('a')
            ket_qua += chr(ma_ky_tu_moi)
        else:
            ket_qua += ky_tu
            
    return ket_qua
tin_nhan_goc = input("Nhập tin nhắn cần giải mã: ")

tin_nhan_da_giai_hoa = giai_ma_caesar(tin_nhan_goc)

print("------------------------------------")
print(f"Tin nhắn gốc: {tin_nhan_goc}")
print(f"Tin nhắn đã mã hóa: {tin_nhan_da_giai_hoa}")
```
