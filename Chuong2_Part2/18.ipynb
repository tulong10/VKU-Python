```python
ch = input (" Nhap vao mot so:")
d = {"chu_hoa": 0, "chu_thuong": 0}
for i in ch:
    if i.isupper()==True:
         d["chu_hoa"] +=1
    elif i.islower() == True:
         d["chu_thuong"] +=1
    else :
        pass
print ("Chu hoa", d["chu_hoa"])
print ("Chu thuong", d["chu_thuong"])
    
```
